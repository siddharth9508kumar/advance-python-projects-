# Exceptio handling

```python
#Q1 handdling the devide by zero without try and except
a=int(input("enter the first number"))
b=int(input("enter the second number"))
if b==0:
    print("devide by zero error")
else:
  print(a/b)
```

```python
#Q2 with the use of try and except
try:
  a=int(input("enter the first number"))
  b=int(input("enter the second number"))
  print(a/b)
  print("done")
except:
  print("can't devide a number by zero")
finally:
  print("Thank you")
```

```python
#Q3 handlling specific exception
try:
  a=int(input("enter the first number"))
  b=int(input("enter the second number"))
  print(a/b)
  print("done")
except:
  print("Something went wrong!!")
finally:
  print("Thank you")
```

```python
#Q4 value error
try:
  a=int(input("enter a number:"))
except ValueError:
  print("invalid input")
else:
  print(a)
```

```python
#Q5 FileNotFound error
try:
  f=open("se.txt")
except FileNotFoundError:
  print("file not found")
f.close()
```

```python
#Q6 Else block
try:
  f=open("sample.txt")
except FileNotFoundError:
  print("file not found")
else:
  print("file created")
f.close()
```

```python
#Q7 finally block
try:
  f=open("sample.txt")
except FileNotFoundError:
  print("file not found")
finally:
  print("finally block")
f.close()
```
